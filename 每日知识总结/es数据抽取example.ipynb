```python
from elasticsearch import Elasticsearch
import pendulum

es = Elasticsearch([
        {"host": "172.16.226.114", "port": 9200}, 
    ],
    http_auth=("admin", "O34uTbYtveak"), 
    use_ssl=True,
    verify_certs=False,
    timeout=3600)
```

```python
end = pendulum.now("Asia/Shanghai")
begin = pendulum.today("Asia/Shanghai")

print(begin, end)
```

```python
query2 = {
    "query":{
        "bool": {
            "must": [
                {
                "range": {
                    "@timestamp": {
                    "format": "strict_date_optional_time",
                    "gte": begin.to_w3c_string(),
                    "lte": end.to_w3c_string()
                    }
                }
                }
            ],
            "filter": [
                {
                "multi_match": {
                    "type": "best_fields",
                    "query": "humbasNo验证不通过",
                    "lenient": True
                }
                }
            ],
            "should": [],
            "must_not": []
        }
    }
}
query = es.search(index="filebeat-*", body=query2)
query['hits']['hits']
```

```python
for hit in query['hits']['hits']:
    print(hit['_source']['message'])
```
